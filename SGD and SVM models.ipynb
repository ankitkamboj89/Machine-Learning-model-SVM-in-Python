Refining the data below, separating the label and explanatory variable columns in two different dataframes

```python
train_data_refined_lables = train_data_refined["class"].copy() #separating the data: attributes from labels
train_data_refined = train_data_refined.drop("class", axis=1) 

test_data_refined_lables = test_data_refined["class"].copy() 
test_data_refined = test_data_refined.drop("class", axis=1) 
```

**Question 4: Standard scaler applied on the data to scale the features on both train and Test data (separately)**

```python
from sklearn.pipeline import Pipeline #applying standscaler to scale the attributes
from sklearn.preprocessing import StandardScaler

std_pipeline = Pipeline([
        ('std_scaler', StandardScaler()),
    ])

train_data_refined_tr = std_pipeline.fit_transform(train_data_refined)
```

```python
test_data_refined_tr = std_pipeline.fit_transform(test_data_refined) #scaling the attributes in test data spearately from the train data usinf same scaler as on train data
```

**Question 5: Fitting the support vector machine classifier**

**Hyperparameters** : <br>
1) kernels: ['rbf', 'poly'] <br>
2) gamma: [.1,2,5] <br>
3) C = [.1,1,100,1000] <br>
4) degree:[1,2,3,4,5,6] <br>
5) Cross-validations: 3 <br>
<br>
**Model Selection**: <br>
1) for kernel 'rbf', accuracy of model is 84% on train data, which is less than 89% accuracy of kernel 'poly'. <br>
2) cross validation score on test data of kernel 'poly' (32%) is lot better than kernel 'rbf' (88%). <br>
<br>
So we select SVM classifier with hyperparameter: 'C': 100, 'degree': 1, 'kernel': 'poly' for futher comparision with SGD classifier.

```python
from sklearn.model_selection import GridSearchCV #training the SVM model for only 'rbf' kernel using grid search and 3 fold cross validation
from sklearn import svm
param_grid={'C':[0.1,1,100,1000], 'kernel' :['rbf'],'gamma':[.1,2,5]}
grid=GridSearchCV(svm.SVC(), param_grid, cv=3, scoring='accuracy')
grid.fit(train_data_refined_tr, train_data_refined_lables)
```

```python
print(grid.best_params_) #getting model with best hyper parameters out of models with 'rbf' kernel
print(grid.score(test_data_refined_tr, test_data_refined_lables))
```

```python
from sklearn.model_selection import GridSearchCV #training the SVM model for only 'poly' kernel using grid search and 3 fold cross validation
param_grid={'C':[0.1,1,100,1000], 'kernel' :['poly'],'degree':[1,2,3,4,5,6]}
grid=GridSearchCV(svm.SVC(), param_grid, cv=3)
grid.fit(train_data_refined_tr, train_data_refined_lables)
```

```python
print(grid.best_params_) #getting model with best hyper parameters out of models with 'poly' kernel
print(grid.score(test_data_refined_tr, test_data_refined_lables))
```

```python
from sklearn.model_selection import cross_val_score #checking cross validation score for 'rbf' kernel models
svc_rbf=svm.SVC(kernel='rbf',C=1, gamma=.1)
svc_rbf.fit(train_data_refined_tr, train_data_refined_lables)
print('Accuracy on test data is:', cross_val_score(svc_rbf,test_data_refined, test_data_refined_lables,cv=3))
print('Accuracy on train data is:', cross_val_score(svc_rbf, train_data_refined_tr, train_data_refined_lables,cv=3))
```

```python
svc_poly1=svm.SVC(kernel='poly',C=100, degree=1) #checking cross validation score for 'poly' kernel models
svc_poly1.fit(train_data_refined_tr, train_data_refined_lables)
print('Accuracy on test data is:', cross_val_score(svc_poly1,test_data_refined, test_data_refined_lables,cv=3))
print('Accuracy on train data is:', cross_val_score(svc_poly1, train_data_refined_tr, train_data_refined_lables,cv=3))
```

**Q6: Stochastic Gradient Descent Classifier**

**Hyperparameters** : <br>
1) 'loss' : ['hinge', 'log', 'modified_huber', 'squared_hinge','perceptron'], <br>
2) 'penalty' : ['l1', 'l2', 'elasticnet'] <br>
<br>
**Model Selection**: <br>
1) SGD model with hyperparameters 'loss': 'squared_hinge', 'penalty': 'elasticnet' with accuracy of 84% on train data.

```python
from sklearn.linear_model import SGDClassifier #training the SGD model
sgd_clf=SGDClassifier(random_state=42)
sgd_clf.fit(train_data_refined_tr, train_data_refined_lables)
```

```python
from sklearn.model_selection import GridSearchCV #using Grid search to find the best model with loss and penality os hyper parameters
param_grid={'loss' : ['hinge', 'log', 'modified_huber', 'squared_hinge','perceptron'], 'penalty' : ['l1', 'l2', 'elasticnet']}
sgd = SGDClassifier(loss="hinge", penalty="l2",random_state=41) 
SGD_Grid = GridSearchCV(sgd, param_grid, cv=3) 
```

```python
SGD_Grid_model = SGD_Grid.fit(train_data_refined_tr, train_data_refined_lables) 
```

```python
print('Best Score: ', SGD_Grid_model.best_score_) #getting the best model from grid search
print('Best Params: ', SGD_Grid_model.best_params_) 
```

```python
from sklearn.model_selection import cross_val_score #checking the accuracy using croos fold score
sgd = SGDClassifier(loss='hinge', penalty='l2',random_state=41) 
sgd_model=sgd.fit(train_data_refined_tr, train_data_refined_lables)
print('Accuracy on test data is:', cross_val_score(sgd_model,test_data_refined, test_data_refined_lables,cv=3))
print('Accuracy on train data is:', cross_val_score(sgd_model, train_data_refined_tr, train_data_refined_lables,cv=3))
```
